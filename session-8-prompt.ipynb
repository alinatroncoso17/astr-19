```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#create an array x ranging from [0,1] in 100 steps

x = np.linspace(0,1,100)
```

```python
#define functions that return sin(x) and cos(x)

def sin_function(x):
    return np.sin(x)

def cos_function(x):
    return np.cos(x)

#create a multipanel plot (1 row, 2 columnns)
#plot sin(x) vs. x in the left panel
#and cos(x) vs. x in the right panel

fig, axes = plt.subplots(1,2)

#first panel sin(x)
axes[0].plot('x,sin_function(x)')
axes[0].set_xlabel('x')
axes[0].set_ylabel('sin(x)')
axes[0].set_title(r'$\sin(x)$')

#second panel cos(x)
axes[1].plot('x,cos_function(x)')
axes[1].set_xlabel('x')
axes[1].set_ylabel('cos(x)')
axes[1].set_title(r'$\cos(x)$')

plt.savefig('sin_cos_plot.pdf')
plt.show()
```
